```python
query = 'phone'
page = 1
webpage = Scraper(f'https://www.amazon.in/s?k={query}&page={page}')
print(webpage.soup.title)
```

```python
# important
target = Tag('div', cls='s-main-slot s-result-list s-search-results sg-row')
items = Tag('div', cls='s-result-item') # repeating products
title = Tag('h2')
price = Tag('span', cls='a-price-whole')
rating = Tag('i', cls='a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom')
```

```python
webpage.get_all(target, items, title=title, price=price, rating=rating)
```

```python
def scrape_product(query, page=1):
    webpage = Scraper(f'https://www.amazon.in/s?k={query}&page={page}')
    print(webpage.soup.title)
    # important
    target = Tag('div', cls='s-main-slot s-result-list s-search-results sg-row')
    items = Tag('div', cls='s-result-item') # repeating products
    title = Tag('h2')
    price = Tag('span', cls='a-price-whole')
    rating = Tag('i', cls='a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom')
    return webpage.get_all(target, items, title=title, price=price, rating=rating)
```

```python
all_data = []
query='laptops'
for i in range(1, 15):
    all_data.extend(scrape_product(query, i))
    print(f'Page {i} done')
```

```python
import pandas as pd
df = pd.DataFrame(all_data).dropna(how='all')
df.to_csv('laptops.csv', index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
filepath = 'laptops.csv'
df = pd.read_csv(filepath)
```

```python
print(df.columns.tolist())
```

```python
def clean_price_column(price):
    if isinstance(price, str):         # check if price variable is a string
        price = price.replace(',', '') # remove comma
        price = price.strip()          # remove whitespace
        if price.isnumeric():
            price = int(price)
            return price
        if len(price) == 0:
            return np.nan
    return price    
```

```python
clean_price_column('23,902')
```

```python
df['cleaned_price'] = df['price'].apply(clean_price_column)
```

```python
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating, str):
        rating = rating.split()[0]
        return float(rating)
    return rating
```

```python
df['cleaned_rating'] = df['rating'].apply(clean_rating_column)
```

```python
df.sort_values(by='cleaned_rating', ascending=False, inplace=True)
df
```
